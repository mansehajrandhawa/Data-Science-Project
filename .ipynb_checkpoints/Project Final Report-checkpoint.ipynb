**Data Description**

This project utilizes two datasets containing player records and session timing data from a University of British Columbia Minecraft server. The players.csv file contains different forms of information used to help identify users. The sessions.csv file contains data on when each user starts and ends their gameplay. Collectively, these datasets contain 11 different variables.
The collection process involved monitoring individuals who connected to UBC's Minecraft server. The server has automated recording capabilities that track when gameplay sessions begin and conclude, with all timestamps logged in UTC (Coordinated Universal Time). The system captures multiple forms of player identification data, such as usernames, age information, and gender details. Email addresses undergo hashing before storage, creating anonymized identifiers for each player.
Direct examination of the players.csv dataset reveals several data quality concerns. Notably, numerous sessions have zero duration. Such entries could result from connection failures or problems with the timestamp recording system. These entries create problems for analysis because they may artificially increase session totals, potentially distorting calculations of concurrent player activity if treated as legitimate sessions. Furthermore, several players showing zero total playing time have somehow achieved pro-level or veteran-level rankings. The apparent inconsistencies between accumulated playtime and status level indicate errors in the reported playtime data.
An additional concern that may not be immediately visible in the data is the possibility of account sharing. When multiple individuals use a single account, data accuracy becomes compromised, especially regarding skill and experience metrics. This occurs because the system attributes all activity from different users to a single account profile.


**Question**

This project's main objective is to understand demand patterns and forecast. We plan to examines which time periods are most probable to experience the highest levels of concurrent player activity. From this general inquiry, we developed a more targeted research question: Is it possible to use the time of day (hour) and weekday to forecast how many gaming sessions begin each hour on the Minecraft server? While this refined question doesn't explicitly measure how many users play simultaneously on the server, it serves as an effective proxy, as hourly session initiation counts correlate strongly with peaks in concurrent gameplay. Peak activity periods are characterized by numerous session start events and considerable session overlap. Our approach will use the sessions.csv dataset to extract the start_time field and convert it into hourly time markers, then compute session counts for each hour. From these hourly markers, we will extract predictor variables indicating the hour of the day and the specific weekday. The result will be a tidy visualization of the hourly-aggregated dataset that is suitable for analysis.

**Hypothesis:** 

Player activity on the Minecraft server will peak during the late night/early morning and evening during the entire week. Deviations from this will be minimal on weekends, when activity will increase. A midday quiet hour will also be represented in the data, as users have other obligations such as work and school. Finally, the hour of the day will be the dominant driver for user concurrency. 

```r
library(readr)
library(dplyr)
library(tidyverse)
library(tidymodels)

options(repr.matrix.max.rows = 6)

player_url   <- "https://raw.githubusercontent.com/mansehajrandhawa/Data-Science-Project/refs/heads/main/players.csv"
sessions_url <- "https://raw.githubusercontent.com/mansehajrandhawa/Data-Science-Project/refs/heads/main/sessions.csv"

players  <- read_csv(player_url)
sessions <- read_csv(sessions_url)

head(players)
cat("Table 1: First six rows of the players dataset\n")

head(sessions)
cat("Table 2: First six rows of the sessions dataset\n")
```

```r
players_wrangled  <- players
sessions_wrangled <- sessions

sessions_wrangled <- sessions_wrangled |>
  mutate(
    # parse start and end times as POSIXct
    start_time_posix = as.POSIXct(start_time,
                                  format = "%d/%m/%Y %H:%M",
                                  tz = "UTC"),
    end_time_posix   = as.POSIXct(end_time,
                                  format = "%d/%m/%Y %H:%M",
                                  tz = "UTC"),
    # fields we actually want for the forecasting question
    date    = as.Date(start_time_posix),
    hour    = as.numeric(format(start_time_posix, "%H")),
    weekday = weekdays(date)
  )


sessions_wrangled |>
  select(hashedEmail, start_time_posix, end_time_posix,
         date, hour, weekday) |>
  head()
cat("Table 3: Example of wrangled session data with date, hour, and weekday\n")
```

```r
hourly_players <- sessions_wrangled |>
  group_by(date, hour, weekday) |>
  summarise(
    unique_players = n_distinct(hashedEmail),
    .groups = "drop"
  )

head(hourly_players)
cat("Table 4: First six rows of hourly unique player counts\n")
```

```r
hourly_summary <- hourly_players |>
  summarise(
    min_players    = min(unique_players,    na.rm = TRUE),
    mean_players   = mean(unique_players,   na.rm = TRUE),
    median_players = median(unique_players, na.rm = TRUE),
    max_players    = max(unique_players,    na.rm = TRUE)
  )

hourly_summary
cat("Table 5: Summary statistics for hourly unique player counts\n")
```

```r
players_by_hour <- hourly_players |>
  group_by(hour) |>
  summarise(
    mean_players = mean(unique_players, na.rm = TRUE),
    .groups = "drop"
  )

players_by_hour
cat("Table 6: Average number of unique players by hour of day\n")


ggplot(players_by_hour,
       aes(x = hour,
           y = mean_players)) +
  geom_col() +
  labs(
    title = "Figure 1: Average number of players by hour of day",
    x = "Hour of day (0–23)",
    y = "Average number of unique players"
  ) +
  theme(
    plot.title = element_text(size = 18, face = "bold")
  )
```

```r
players_by_weekday <- hourly_players |>
  group_by(weekday) |>
  summarise(
    mean_players = mean(unique_players, na.rm = TRUE),
    .groups = "drop"
  )

players_by_weekday
cat("Table 7: Average number of unique players by day of week\n")

ggplot(players_by_weekday,
       aes(x = reorder(weekday, mean_players),
           y = mean_players)) +
  geom_col() +
  coord_flip() +
  labs(
    title = "Figure 2: Average number of players by day of week",
    x = "Day of week",
    y = "Average number of unique players"
  ) +
  theme(
    plot.title = element_text(size = 18, face = "bold")
  )
```

```r
sessions_by_weekday <- sessions_wrangled |>
  group_by(weekday) |>
  summarise(number_of_sessions = n())

sessions_by_weekday
cat("Table 8: Number of sessions by day of week\n")

ggplot(sessions_by_weekday,
       aes(x = reorder(weekday, number_of_sessions),
           y = number_of_sessions)) +
  geom_col() +
  coord_flip() +
  labs(
      title = "Figure 3: Number of sessions by day of week",
      x = "Day of week",
      y = "Number of sessions"
    ) +
  theme(
    plot.title = element_text(size = 18, face = "bold")
  )

```

```r
# Data for high vs low load

hourly_model <- hourly_players |>
  mutate(
    is_weekend = if_else(weekday %in% c("Saturday", "Sunday"), 1, 0),
    high_load  = if_else(unique_players >= 2, "high", "low"),
    high_load  = factor(high_load)
  ) |>
  select(high_load, hour, is_weekend)

head(hourly_model)
cat("Table 9: Example of modelling data with high/low load labels\n")
```

```r
# Set the seed. Don't remove this!
set.seed(9999)

# Define "high load" as the top 25% of hours by unique_players
threshold <- quantile(hourly_players$unique_players,
                      probs = 0.75,
                      na.rm = TRUE)

threshold

hourly_model <- hourly_players |>
  mutate(
    is_weekend = if_else(weekday %in% c("Saturday", "Sunday"), 1, 0),
    high_load  = if_else(unique_players >= threshold, "high", "low"),
    high_load  = as.factor(high_load)
  ) |>
  select(high_load, hour, is_weekend, weekday)

head(hourly_model)
cat("Table 10: Example rows of modelling data with high/low load labels\n")
```

```r
load_recipe <- recipe(high_load ~ ., data = hourly_model) |>
  update_role(high_load, new_role = "outcome") |>
  step_center(all_predictors()) |>
  step_scale(all_predictors())

load_recipe
```

```r
n_rows     <- nrow(hourly_model)
train_size <- floor(0.8 * n_rows)

train_indices <- sample(1:n_rows, size = train_size)

train_data <- hourly_model[train_indices, ]
test_data  <- hourly_model[-train_indices, ]

nrow(train_data)
nrow(test_data)
```

```r
# Scale predictors using training set mean and standard deviation

hour_mean_train <- mean(train_data$hour,       na.rm = TRUE)
hour_sd_train   <- sd(train_data$hour,         na.rm = TRUE)

week_mean_train <- mean(train_data$is_weekend, na.rm = TRUE)
week_sd_train   <- sd(train_data$is_weekend,   na.rm = TRUE)

train_X <- train_data |>
  mutate(
    hour_scaled    = (hour       - hour_mean_train) / hour_sd_train,
    weekend_scaled = (is_weekend - week_mean_train) / week_sd_train
  ) |>
  select(hour_scaled, weekend_scaled)

test_X <- test_data |>
  mutate(
    hour_scaled    = (hour       - hour_mean_train) / hour_sd_train,
    weekend_scaled = (is_weekend - week_mean_train) / week_sd_train
  ) |>
  select(hour_scaled, weekend_scaled)

train_y <- train_data$high_load
test_y  <- test_data$high_load

head(train_X)
cat("Table 11: Scaled predictors for training set (train_X)\n")

head(test_X)
cat("Table 12: Scaled predictors for test set (test_X)\n")

```

```r
knn_pred <- class::knn(
  train = as.matrix(train_X),
  test  = as.matrix(test_X),
  cl    = train_y,
  k     = 7
)

knn_accuracy <- mean(knn_pred == test_y)
knn_accuracy

table(predicted = knn_pred, actual = test_y)
cat("Table 13: Matrix for the k-NN model on the test set\n")
```

```r
test_results <- test_data |>
  mutate(
    predicted_high = knn_pred
  )

head(test_results)
cat("Table 14: Test data with predicted high-load labels\n")
```

```r
pred_by_hour <- test_results |>
  group_by(hour) |>
  summarise(
    prop_pred_high = mean(predicted_high == "high", na.rm = TRUE),
    .groups = "drop"
  )

pred_by_hour
cat("Table 15: Proportion of hours predicted high load by hour of day\n")

ggplot(pred_by_hour,
       aes(x = hour,
           y = prop_pred_high)) +
  geom_col() +
  labs(
    title = "Figure 4: Proportion of hours predicted as high load",
    x = "Hour of day (0–23)",
    y = "Proportion of hours predicted high load"
  ) +
  theme(
    plot.title = element_text(size = 18, face = "bold")
  )

```

```r
pred_by_weekday <- test_results |>
  group_by(weekday) |>
  summarise(
    prop_pred_high = mean(predicted_high == "high", na.rm = TRUE),
    .groups = "drop"
  )

pred_by_weekday
cat("Table 16: Proportion of hours predicted high load by weekday\n")

ggplot(pred_by_weekday,
       aes(x = reorder(weekday, prop_pred_high),
           y = prop_pred_high)) +
  geom_col() +
  coord_flip() +
  labs(
  title = "Figure 5: Proportion of hours predicted as high load by weekday",
    x = "Day of week",
    y = "Proportion of hours predicted high load"
  ) +
  theme(
    plot.title = element_text(size = 18, face = "bold")
  )

```
